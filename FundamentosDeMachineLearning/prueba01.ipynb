```python
# Importar libreras
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, norm

# Funcin para graficar distribuciones
def graficar_distribucion(x_vals, pmf_vals, datos, titulo, xlabel, ylabel, color, bins=None):
    plt.figure(figsize=(8, 5))
    if bins is not None:
        plt.hist(datos, bins=bins, density=True, alpha=0.7, color=color, edgecolor='black', label="Histograma")
    plt.plot(x_vals, pmf_vals, 'o-', color="red", label="Distribucin terica")
    plt.title(titulo)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True, alpha=0.5)
    plt.show()

# Distribucin Binomial
def distribucion_binomial():
    p = 0.3  # probabilidad de xito
    n = 10  # nmero de ensayos
    num_muestras = 10000

    # Generar datos simulados
    datos = np.random.binomial(n, p, num_muestras)
    k_valores = np.arange(0, n + 1)
    pmf_valores = binom.pmf(k_valores, n, p)

    # Graficar
    graficar_distribucion(
        k_valores, pmf_valores, datos,
        titulo=f'Distribucin Binomial (n={n}, p={p})',
        xlabel="Nmero de xitos",
        ylabel="Probabilidad",
        color="#34495E",
        bins=np.arange(-0.5, n + 1.5)
    )

# Distribucin de Poisson
def distribucion_poisson():
    lambda_poisson = 4  # razn media
    num_muestras = 10000

    # Generar datos simulados
    datos = np.random.poisson(lam=lambda_poisson, size=num_muestras)
    k_vals = np.arange(0, max(datos) + 1)
    pmf_vals = poisson.pmf(k_vals, lambda_poisson)

    # Graficar
    graficar_distribucion(
        k_vals, pmf_vals, datos,
        titulo=f'Distribucin de Poisson (={lambda_poisson})',
        xlabel="Nmero de ocurrencias (k)",
        ylabel="Probabilidad",
        color="#AEB6BF",
        bins=k_vals
    )

# Distribucin Normal
def distribucion_normal():
    media = 10
    desviacion = 1
    num_muestras = 10000

    # Generar datos simulados
    datos_sim = np.random.normal(media, desviacion, num_muestras)
    x_vals = np.linspace(min(datos_sim), max(datos_sim), 300)
    pdf_vals = norm.pdf(x_vals, media, desviacion)

    # Graficar
    plt.figure(figsize=(8, 5))
    plt.hist(datos_sim, bins=50, density=True, alpha=0.7, color='#1B2631', edgecolor='white', label="Histograma")
    plt.plot(x_vals, pdf_vals, 'r-', linewidth=3, label="Distribucin Normal")
    plt.title(f'Distribucin Normal (media={media}, desviacin estndar={desviacion})')
    plt.xlabel("Valores de x")
    plt.ylabel("Densidad de probabilidad")
    plt.legend()
    plt.grid(True, alpha=0.5)
    plt.show()

# Llamar a las funciones
distribucion_binomial()
distribucion_poisson()
distribucion_normal()
```

```python
# Distribucin de Poisson
def graficar_poisson(valores_lambda, num_muestras, colores):
    """
    Genera histogramas y curvas tericas para distribuciones de Poisson con diferentes valores de lambda.

    :param valores_lambda: Lista de valores de lambda (razn media).
    :param num_muestras: Nmero de muestras a generar.
    :param colores: Lista de colores para cada lambda.
    """
    plt.figure(figsize=(9, 4))
    for lam, color in zip(valores_lambda, colores):
        # Generar datos simulados
        datos = np.random.poisson(lam, num_muestras)
        k_vals = np.arange(0, max(datos) + 1)  # Valores de k
        pmf_vals = poisson.pmf(k_vals, lam)   # Probabilidades tericas

        # Graficar histograma y curva terica
        plt.hist(datos, bins=np.arange(-0.5, max(datos) + 1.5), density=True, alpha=0.4, color=color, edgecolor='black', label=f'={lam}')
        plt.plot(k_vals, pmf_vals, color=color, linewidth=2)

    # Personalizar grfica
    plt.title('Histogramas de Poisson para diferentes valores de ')
    plt.xlabel('Nmero de ocurrencias (k)')
    plt.ylabel('Probabilidad')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
    # Parmetros de la distribucin
valores_lambda = [5, 9, 15]  # Razn media
num_muestras = 10000
colores = ['#82E0AA', '#FAE5D3', '#95A5A6']

# Llamar a la funcin
graficar_poisson(valores_lambda, num_muestras, colores)
```

```python


def graficar_distribucion_normal(media, desviacion, num_muestras, bins=50):
    """
    Genera un histograma y la curva terica de una distribucin normal.

    :param media: Media de la distribucin normal.
    :param desviacion: Desviacin estndar de la distribucin normal.
    :param num_muestras: Nmero de muestras a generar.
    :param bins: Nmero de bins para el histograma.
    
    """
    # Generar datos simulados
    datos_sim = np.random.normal(media, desviacion, num_muestras)

    # Generar histograma
    conteo, bins, _ = plt.hist(datos_sim, bins=bins, density=True, alpha=0.7, color='#1B2631', edgecolor='white', label="Histograma")

    # Calcular la PDF terica
    x_vals = np.linspace(min(bins), max(bins), 300)
    pdf_vals = norm.pdf(x_vals, media, desviacion)

    # Graficar la curva terica
    plt.plot(x_vals, pdf_vals, 'r-', linewidth=3, label="Distribucin Normal")
    plt.title(f'Distribucin Normal (media={media}, desviacin estndar={desviacion})')
    plt.xlabel("Valores de x")
    plt.ylabel("Densidad de probabilidad")
    plt.legend()
    plt.grid(True, alpha=0.5)
    plt.show()

# Parmetros de la distribucin normal
media = 10
desviacion = 1
num_muestras = 10000

# Llamar a la funcin
graficar_distribucion_normal(media, desviacion, num_muestras)
```
